```python
import pandas as pd
df = pd.read_csv('C:\\Users\\USER\\Documents\\Dir-S\\Paris_Olympics\\Cleaned_Dataset\\medals.csv')
#Creating rank column based on Winning Gold medals
df['rank'] = df['Gold Medal'].rank(ascending=False, method='min').astype(int)
#Creating Rank by total column based of total number of medals Won
df['rank by total'] = df['Total'].rank(ascending=False, method='min').astype(int)
df.head()
output_path = r'C:\Users\USER\Documents\Dir-S\Paris_Olympics\Cleaned_Dataset\medalsbytotal.csv'
df.to_csv(output_path, index=False)

```
